```python
from PIL import Image, ImageFile
from matplotlib.pyplot import imshow
import requests
from io import BytesIO
from sklearn import metrics
import numpy as np
import pandas as pd
import tensorflow as tf
from IPython.display import display, HTML 

images = [
    "https://upload.wikimedia.org/wikipedia/commons/a/a1/Tartu_Toomkiriku_varemed_2012.jpg",
    "https://upload.wikimedia.org/wikipedia/commons/2/29/%D0%92%D0%BE%D1%80%D0%BE%D1%82%D0%B0_%D0%B2_%D0%9F%D0%B5%D1%82%D1%80%D0%BE%D0%BF%D0%B0%D0%B2%D0%BB%D0%BE%D0%B2%D1%81%D0%BA%D1%83%D1%8E_%D1%86%D0%B5%D1%80%D0%BA%D0%BE%D0%B2%D1%8C_%28%D0%9C%D0%BE%D1%81%D0%BA%D0%BE%D0%B2%D1%81%D0%BA%D0%B0%D1%8F_%D0%BE%D0%B1%D0%BB._%D0%BF%D0%BE%D1%81._%D0%98%D0%BB%D1%8C%D0%B8%D0%BD%D1%81%D0%BA%D0%BE%D0%B5%29.jpg",
    "https://upload.wikimedia.org/wikipedia/commons/6/60/Statsbiblioteket_l%C3%A6sesalen-2.jpg",
    "https://upload.wikimedia.org/wikipedia/commons/9/92/Brookings.jpg",
    "https://upload.wikimedia.org/wikipedia/commons/f/ff/" +  "WashU_Graham_Chapel.JPG",
    "https://upload.wikimedia.org/wikipedia/commons/9/9e/SeigleHall.jpg",
    "https://upload.wikimedia.org/wikipedia/commons/2/22/Maastricht_Vrijthof_15_BW_2017-08-19_12-06-24.jpg",
    "https://upload.wikimedia.org/wikipedia/commons/0/00/Labastide-Saint-Sernin_-_L%27%C3%A9glise.jpg",
    "https://upload.wikimedia.org/wikipedia/commons/b/b5/N%C3%B6tsch_Saak_Pfarrkirche_hl_Kanzian_08052015_3382.jpg",
    "https://upload.wikimedia.org/wikipedia/commons/a/aa/WUSTLKnight.jpg",
    "https://upload.wikimedia.org/wikipedia/commons/3/32/WashUABhall.jpg",
    "https://upload.wikimedia.org/wikipedia/commons/c/c0/Brown_Hall.jpg",
    "https://upload.wikimedia.org/wikipedia/commons/f/f4/South40.jpg", 
    "https://upload.wikimedia.org/wikipedia/commons/9/94/Balcones_de_la_Avenida_Maritima_-_Santa_Cruz_de_La_Palma_14.jpg",
    "https://upload.wikimedia.org/wikipedia/commons/0/00/Labastide-Saint-Sernin_-_L%27%C3%A9glise.jpg"
]


def make_square(img):
    cols,rows = img.size
    
    if rows>cols:
        pad = (rows-cols)/2
        img = img.crop((pad,0,cols,cols))
    else:
        pad = (cols-rows)/2
        img = img.crop((0,pad,rows,rows))
    
    return img

headers = {
    "user-agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.141 Safari/537.36"
}

x = []    
y = []
loaded_images = []

for url in images:
    response = requests.get(url,headers=headers)
    img = Image.open(BytesIO(response.content))
    img.load()
    img = make_square(img)
    img = img.resize((128,128), Image.ANTIALIAS)
    
    loaded_images.append(img)
    print(url)
    display(img)
    for i in range(20):
        img_array = np.asarray(img)
        img_array_noise = add_noise(img_array)
        
        img_array = img_array.flatten()
        img_array = img_array.astype(np.float32)
        img_array = (img_array-128)/128
        
        img_array_noise = img_array_noise.flatten()
        img_array_noise = img_array_noise.astype(np.float32)
        img_array_noise = (img_array_noise-128)/128
        
        x.append(img_array_noise)
        y.append(img_array)

x = np.array(x)
print(x)
y = np.array(y)
print(y)

```
